```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.graph_objs as go
```

```python
def generate_random(dist="norm", count = 0):
    if dist == 'norm':
        return np.random.normal(size = count), np.random.normal(size = count)
    elif dist == 'unif':
        return np.random.uniform(size = count), np.random.uniform(size = count)
```

```python
x, y = generate_random('norm', 10)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line plot')
plt.show()
```

```python
x, y = generate_random('unif', 10)
```

```python
plt.plot(x, y)
```

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot')
plt.show()
```

```python
plt.boxplot(x)
plt.xlabel('x')
plt.title('Box plot for x data')
plt.show()
```

```python
plt.boxplot(y)
plt.xlabel('x')
plt.title('Box plot for y data')
plt.show()
```

```python
sns.boxplot(y)
```

```python
# Creating trace1
trace1 = go.Scatter(
                    x = x)
# Creating trace2
trace2 = go.Scatter(
                    x = y)
data = [trace1, trace2]
layout = dict(title = 'Citation and Teaching vs World Rank of Top 100 Universities',
              xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
go.iplot(fig)
```
